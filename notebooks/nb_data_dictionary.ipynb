# Load Bronze Dataset

**Purpose:**    
Load the raw Bronze dataset to inspect its schema and document every column before designing the Silver layer.

```python
accepted_df = (
    spark.read
         .option("header", "true")
         .option("inferSchema", "true")
         .csv("Files/bronze/accepted_loans/accepted_2007_to_2018Q4.csv.gz")
)
```

```python
columns_df = spark.createDataFrame(
    [(c,) for c in accepted_df.columns],
    ["Column Name"]
)

display(columns_df)
```

```python
from pyspark.sql.types import *

data_dictionary = [
    (
        i + 1,
        field.name,
        str(field.dataType).replace("Type()", ""),
        "Yes" if field.nullable else "No"
    )
    for i, field in enumerate(accepted_df.schema.fields)
]

dictionary_df = spark.createDataFrame(
    data_dictionary,
    ["#", "Column Name", "Data Type", "Nullable"]
)

display(dictionary_df)
```
